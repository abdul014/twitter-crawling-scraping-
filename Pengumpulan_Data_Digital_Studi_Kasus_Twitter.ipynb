### Web Scraping Ethics

* Scrapping secara umum (minimal) tidak etis (bisa jadi ilegal). Mengapa? karena scraping
mirip DDOS attack yang akan memberatkan server atau bahkan membuat server berhenti
berfungsi normal.
*Program scraping juga memungkinkan akan mengakses data yang tidak dimaksudkan untuk konsumsi publik.
Satu-satunya saat dimana scraping boleh dilakukan adalah saat sang pelaku/programer
menghormati ‚Äúrobots.txt‚Äù yang telah diberikan oleh web administrator. Atau lebih baik lagi
adalah menggunakan API (Application Program Interface) yang diberikan oleh provider
(website/medsos) lalu melakukan crawling.
* Yakinkan untuk membaca ToS (Terms of Sevice) dengan baik.
* **More details here:** https://tau-data.id/scraping

### Menyimpan dictionary ke json

````python
import json
def saveTweets(tweets, file='Tweets.json'): #in Json Format
 with open(file, 'w') as f:
  for t in tweets:
   try:
    f.write(json.dumps(t)+'\n')
   except:
    pass
````

````python
saveTweets(tweets, file=filename+".json")
````

### Load json to Dictionary

````python
def loadTweets(file='Tweets.json'):
 f=open(file,encoding='utf-8', errors ='ignore', mode='r')
 T=f.readlines();f.close()
 for i,t in enumerate(T):
   T[i] = json.loads(t.strip())
 return T
````

````python
# Atau bisa coba dengan mengirim tweet sendiri :)
from twython import TwythonStreamer

def streamTwitter(topicS, lang):
    class MyStreamer(TwythonStreamer):
        def on_success(self, data):
            global count
            count+=1
            print('tweet from {}, post: {}'.format(data['user']['screen_name'], data['text']))
            if count==maxTweet:
                print('\nFinished streaming %.0f tweets' %(maxTweet)); self.disconnect()

        def on_error(self, status_code, data):
            print('Error Status = %s' %status_code); self.disconnect()
    while count<maxTweet:
        stream = MyStreamer(consumer_key,consumer_key_secret,access_token,access_token_secret)
        stream.statuses.filter(track=topicS)
````

````python
maxTweet, count = 7, 0 # Rubah sesuai dengan kebutuhan, Untuk percobaan ini cukup (misal) 3 tweet
lang = set(['en','id']) # bahasa bisa dipilih > 1
topicS = ['KDRT'] # Bisa>1

streamTwitter(topicS, lang)
````
