# **Akuisisi Data Teks**

Akuisisi data teks adalah proses pengumpulan, pengambilan, atau pengumpulan informasi dalam bentuk teks dari berbagai sumber. Tujuan akuisisi data teks dapat bervariasi, termasuk analisis sentimen, pemrosesan bahasa alami, klasifikasi dokumen, dan berbagai aplikasi lainnya yang melibatkan pemahaman dan ekstraksi informasi dari teks.

Beberapa jenis data:
1. Data Primer
  - Survey, Sensus
  - Eksperimen/Percobaan
  - Scraping, Crawling
2. Data Skunder : data yang bersumber dari pihak ketiga

### **Crawling vs Scraping**
Web Crawling vs Web Scraping 2.png.webp

Sumber: [SOAX](https://soax.com/blog/web-crawling-vs-web-scraping-what-is-the-difference)

**Web Crawling:**
- Definisi: Web crawling adalah proses secara otomatis mengumpulkan informasi dari berbagai situs web dengan menjelajahi struktur hyperlink antarhalaman.
- Tujuan: Biasanya digunakan oleh mesin pencari untuk memperbarui indeks mereka dan menemukan halaman baru.
- Cara Kerja: Sebuah bot web (spider atau crawler) melewati berbagai situs web, mengikuti tautan, dan mengindeks konten halaman web tanpa memperhatikan struktur atau format spesifik dari data.

**Web Scraping:**
- Definisi: Web scraping adalah proses mengekstrak data tertentu dari halaman web dengan menggunakan program atau skrip.
- Tujuan: Digunakan untuk mengambil informasi tertentu yang diinginkan dari halaman web, seperti harga produk, ulasan pengguna, atau data lainnya.
- Cara Kerja: Sebuah skrip atau program dibuat untuk mengakses halaman web, mengekstrak data tertentu, dan menyimpannya dalam format yang dapat diolah.

**Perbedaan utama:**
- Fokus: Web crawling lebih berfokus pada mengindeks dan menemukan halaman-halaman baru di internet, sedangkan web scraping berfokus pada ekstraksi data spesifik dari halaman web.
- Orientasi: Web crawling bersifat lebih umum dan bersifat luas, sedangkan web scraping lebih fokus pada informasi tertentu atau target spesifik.
- Izin: Web crawling umumnya lebih bersifat terbuka karena dilakukan oleh mesin pencari dan agen pengindeks, sedangkan web scraping dapat menimbulkan masalah legal karena dapat dianggap sebagai pengambilan data tanpa izin dari pemilik situs web.

### ***Legality Issue***

1. Urusan hak cipta
  - Situs web umumnya dilindungi oleh hak cipta, dan penggunaan otomatis atau pengambilan konten mereka tanpa izin dapat dianggap sebagai pelanggaran hak cipta.
  - Beberapa situs web memiliki file robots.txt yang memberikan petunjuk kepada web crawlers tentang bagian mana dari situs yang boleh atau tidak boleh diindeks. Melanggar aturan ini dapat mengakibatkan masalah hukum.
2. Pelanggaran Aturan Penggunaan Situs Web
  - Banyak situs web memiliki ketentuan penggunaan atau perjanjian lisensi pengguna akhir yang melarang scraping atau penggunaan otomatis tanpa izin. Melanggar ketentuan ini dapat menimbulkan konsekuensi hukum.
3. Privasi
  - Web scraping yang melibatkan pengumpulan informasi pribadi tanpa izin dapat melanggar undang-undang privasi, terutama jika informasi tersebut digunakan tanpa izin atau dengan cara yang melanggar aturan privasi.
4. Penggunaan Kapasitas Server:
  - Web crawling yang dilakukan dalam skala besar dan tanpa izin dapat menyebabkan beban server yang berlebihan, yang dapat dianggap sebagai serangan terhadap server. Ini bisa melanggar undang-undang yang melarang serangan terhadap sistem komputer.

```python
import requests
import twarc.expansions as expansions
import pandas as pd
import os

search_url = "https://api.twitter.com/2/tweets/search/recent"
bearer_token = userdata.get("bearer_token")
```

```python
def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token), "User-Agent": "v2UserTweetsPython"}
    return headers


def connect_to_endpoint(url, headers, params):
    response = requests.request("GET", search_url, headers=headers, params=params)

    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
query_params = {'query': 'jokowi lang:id -is:retweet -is:reply -has:mentions -has:links',
            'tweet.fields':'created_at,lang,text,geo,author_id,id,public_metrics,referenced_tweets',
            'expansions':'geo.place_id,author_id',
            'place.fields':'contained_within,country,country_code,full_name,geo,id,name,place_type',
            'user.fields':'description,username,id',
            'start_time':'2024-01-01T00:00:01.000Z',
            'end_time':'2024-01-31T23:59:00.000Z',
            'max_results':'100'}
```

```python
# membuat otorisasi dengan bearer token API
headers = create_headers(bearer_token)
```

```python
# melakukan koneksi ke twitter API dengan bearer token
json_response = connect_to_endpoint(search_url, headers, query_params)
```

##**Akuisisi Data Shopee**

Perlu digaris bawahi bahwa contoh ini ***ilegal*** karena Shopee tidak membenarkan pengambilan data _rating_ pada `robots.txt` mereka, dapat dilihat di link [ini](https://shopee.co.id/robots.txt).

Sehingga, jika melakukan riset pada data Shopee, usahakan terlebih dahulu mendaftar di platform khusus developer mereka.

```python
import re
import requests
import pandas as pd
```

Misalkan dilakukan pengambilan data pada web berikut:

```python
url = 'https://shopee.co.id/Apple-iPhone-13-128GB-Green-i.241308147.17430259949?xptdk=1896514d-037c-42ca-a9d9-4db179403fc8'
```

```python
r = re.search(r'i\.(\d+)\.(\d+)', url)
shop_id, item_id = r[1], r[2]
ratings_url = 'https://shopee.co.id/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
```

```python
offset = 0
d = {'username': [], 'rating': [], 'comment': []}
while True:

    data = requests.get(ratings_url.format(shop_id=shop_id, item_id=item_id, offset=offset)).json()

    # uncomment ini untuk menampilkan seluruh data
    # print(json.dumps(data, indent=4))

    i = 1
    for i, rating in enumerate(data['data']['ratings'], 1):
        d['username'].append(rating['author_username'])
        d['rating'].append(rating['rating_star'])
        d['comment'].append(rating['comment'])

        print(rating['author_username'])
        print(rating['rating_star'])
        print(rating['comment'])
        print('-' * 80)

    if i % 20:
        break

    offset += 20
```

```python
df = pd.DataFrame(d)
df[df['comment'] != ""]['comment'][4]
# df.to_csv('/content/drive/MyDrive/Asistensi/Materi TA/Pertemuan 2/data_shopee.csv', index=False)
```

##**Akuisisi Data Berita**

Pada akuisisi data berita kali ini, akan digunakan dua _package_ untuk melakukan _crawling_ data dengan bantuan NLP, yakni menggunakan `GoogleNews` dan `newspaper3k`. Selain dua _package_ ini, ada _package_ lain yang bisa digunakan seperti `PyGoogleNews` dan `NewsCatcher`.

```python
from GoogleNews import GoogleNews
from newspaper import Article
import pandas as pd
```

```python
search_term = "ria ricis" #keyword

googlenews=GoogleNews(start='01/01/2024',end='01/02/2024', lang='id', region='ID')
googlenews.search(search_term)
```

```python
results = []

for page in range(1,3): #dari page 1 sampai 2
  googlenews.getpage(page)
  results.extend(googlenews.result())
```

```python
results
```

```python
df_news = pd.DataFrame(results)
df_news
```

Untuk melihat apa saja platform media yang terambil:

```python
df_news['media'].unique()
```

Membuat fungsi untuk mendapatkan isi/konten dari masing-masing link berita:

```python
from newspaper import Article

def printNews (url):
    toi_article = Article(url, language="id") # en for English
    try:
        #Download artikel
        toi_article.download()

        #Parse artikel
        toi_article.parse()

        #NLP artikel
        toi_article.nlp()

    except:
        pass

    #To extract title
    #print("Article's Title:")
    #print(toi_article.title)
    #print()

    #To extract text
    #print("Article's Text:")
    #print(toi_article.text)
    #print()

    #To extract summary
    # print("Article's Summary:")
    # print(toi_article.summary)
    # print()

    #To extract keywords
    # print("Article's Keywords:")
    return(toi_article.text)
```

Simpan berita dalam bentuk dataframe

```python
temp = []
for i in range(0, len(df_news)):
    temp.append({
        'title': df_news['title'][i],
        'date' : df_news['date'][i],
        'media' : df_news['media'][i],
        'link' : df_news['link'][i],
        'text' : printNews(df_news['link'][i])
    })
    print(i) # melihat indeks yang berjalan
news = pd.DataFrame(temp)
```

```python
news
```

```python
# menghapus data kosong
empty_id = news[news['text']==""].index
news.drop(empty_id, inplace = True)
news = news.reset_index()
news
```

```python
news = news[['title', 'date',	'media',	'link',	'text']]
news
```
